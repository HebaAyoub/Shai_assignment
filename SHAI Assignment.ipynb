## Handling columns that are not completely Null

### Check the percentage of null value in each column to dealing with it 

```python
for column in df.columns:
    null_count = df[column].isnull().sum()*100/len(df)
    print (f"the percentage of Null values in column {column} is: {null_count}")
```

#### Fill the Null values in columns BasePay, OvertimePay, OtherPay with mean of these columns as the percentage of this null data is verry small and not affect.

#### According to the coulomn Benefits as the percentage of null values are aprroximately a quarter of the total data, the mean value The average will be somewhat far from the true value

```python
lst = ["BasePay", "OvertimePay", "OtherPay"]
for column in lst:
    df[column].fillna(df[column].mean(), inplace=True)
df["Benefits"].fillna(0, inplace=True)
```

## Drawing Pie chart to represent the proportion of employees based on their departements

### Extract department from Job title

```python
def extract_department(job_title):
    if 'engineer' in job_title.lower():
        return 'Engineering'
    elif 'manager' in job_title.lower():
        return 'Management'
    elif 'police' in job_title.lower():
        return 'Police'
    elif 'fire' in job_title.lower():
        return 'Fire'
    elif 'account' in job_title.lower():
        return 'Accounting'
    else:
        return 'Other'
df['JobTitle'] = df['JobTitle'].astype(str)
df['Department'] = df['JobTitle'].apply(extract_department)
```

```python
department_counts = df['Department'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

## Grouped Analysis

```python
summary_stats = df.groupby(['Department', 'Year']).agg({
    'TotalPay': 'mean', })
print(summary_stats)

average_salary_by_department = df.groupby('Department')['TotalPay'].mean()
print("\nAverage salaries by department:")
print(average_salary_by_department)
```

```python
correlation = df['TotalPay'].corr(df['Year'])
print("Correlation coefficient between 'TotalPay' and 'Year':", correlation)
# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['Year'], df['TotalPay'], alpha=0.5)
plt.title('Scatter Plot: TotalPay vs Year')
plt.xlabel('Year')
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```

```python
correlation = df['TotalPay'].corr(df['BasePay'])
print("Correlation coefficient between 'TotalPay' and 'BsePay':", correlation)
# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['TotalPay'], df['BasePay'], alpha=0.5)
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```
